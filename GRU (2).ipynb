```python
def remove_nan(column_value):
    for i in range(len(column_value)):
        if(np.isnan(column_value[i])==True):
            column_value[i]=column_value[i-1] #replacing by the previous value of stock as averaging would not be beneficial in our case
    return column_value
```

```python
df=remove_nan(df)
```

```python
print(df.isnull().sum())
```

```python
df=np.array(df)
training_size=int((0.7)*len(df))
validation_size=int((0.2)*len(df))
```

```python
#splitting data into 70:20:10 train:validation:test ratio
training_data=df[:training_size]
validation_data=df[training_size:training_size+validation_size]
testing_data=df[training_size+validation_size:]
```

```python
training=[training_data.max(),training_data.min()]
validation=[validation_data.max(),validation_data.min()]
testing=[testing_data.max(),testing_data.min()]
def transform(column_value):
    #changing to range [0,1]
    return (column_value-column_value.min())/(column_value.max()-column_value.min())
```

```python
def inverse_transform(scaled_column_values,max_min):
    maximum=max_min[0]
    minimum=max_min[1]
    return scaled_column_values*(maximum-minimum)+minimum
```

```python
def split(data,span=60):
    x=[]
    y=[]
    for i in range(len(data)-span):
        x.append(data[i:i+span])
        y.append(data[i+span])
    x=np.array(x)
    y=np.array(y)
    return (x,y)
```

```python
#creating split training_data
x_train,y_train=split(training_data,50)
print(x_train.shape)
print(y_train.shape)
```

```python
#creating split validation_data
x_validation,y_validation=split(validation_data,50)
print(x_validation.shape)
print(y_validation.shape)
```

```python
#creating split testing_data
x_testing,y_testing=split(testing_data,50)
print(x_testing.shape)
print(y_testing.shape)
```
