```python
train=np.vstack((Asian,Axis,Bajaj,Baj,Bharti,Bpcl,Bri,Cipla,Bajaj_fin))
train.shape
```

```python
columns=['Date','Open','High','Low','Close','Adj Close','volume','code']
training_data=pd.DataFrame(train, columns = columns)
#training_data #this is our training data
```

```python
#feature scaling on the column Open(our output) so that the features don't dominate over other features
#want to get rid of biases in the data
from sklearn.preprocessing import MinMaxScaler
train_feature_scaling=(training_data.iloc[:,1].values).reshape(-1,1)
ss= MinMaxScaler(feature_range=(0,1))
train_open_scaled= ss.fit_transform(train_feature_scaling)
```

```python
train_open_scaled[100]
```

```python
#so 60 timestamps data created
x_train=[]
y_train=[]
for i in range(60,len(train_open_scaled)):
    x_train.append(train_open_scaled[i-60:i,0]) #so each time i-60 to i-1 th row is our input and ith row will be our output ie predict the 60th days result based on previous 60 days
    y_train.append(train_open_scaled[i,0])
#so x_train array has each element as record of what open price of stock was 60 days before
#so x_train[3]=> list of open price of stock from day 3 to day 62 and corresponding y_train[3] would have open price of stock at day 63
x_train = np.array(x_train)
y_train= np.array(y_train)
```

```python
#Reshaping the train data to make it as input for LTSM layer input_shape(batchzise,timesteps,input_dim)
print(x_train.shape)
x_train= np.reshape(x_train,(x_train.shape[0],x_train.shape[1],1))
print(x_train.shape) #as needed an in layer value 1 needed to add as LSTM look at this like layers
```

```python
x_train.shape
```

```python
reg=Sequential() #reg is object of class Sequential, so initialised our RNN 
```

```python
#First Input layer and LSTM layer with 0.2% dropout and kernel_initializer as random_normal
#droupout would drop unnessecary datas
reg.add(LSTM(units=50,return_sequences=True,kernel_initializer='random_normal',input_shape=(x_train.shape[1],1))) #adding 1st LSTM layer here
#so 50 is the dimension of output space ie. 60 coming in and 50 going out of layer
reg.add(Dropout(0.2)) #droupout layer so avoid overfitting turn 20% random neurons off
```

```python
# Second LSTM layer also with 0.2% dropout
reg.add(LSTM(units=50,kernel_initializer='random_normal',return_sequences=True))
reg.add(Dropout(0.2))
```

```python
#Third LSTM layer with 0.2% dropout
reg.add(LSTM(units=50,kernel_initializer='random_normal',return_sequences=True))
reg.add(Dropout(0.2))
```

```python
#Fourth LSTM layer with 0.2% dropout
reg.add(LSTM(units=50,kernel_initializer='random_normal'))
reg.add(Dropout(0.2))
```

```python
#adding Output layer/dense layer
reg.add(Dense(units=1))
#Now we have to Compile the network and fit it
reg.compile(optimizer='adam',loss='mean_squared_error')
```

```python
reg.fit(x_train,y_train,batch_size=32,epochs=100)
```
